### Was the ratio of drug use for the 18-25 age group statistically different from the ratios of use for the other two age groups?
It was determined from Figure 1 that 18-25 year olds had the highest rate of substance use in 2018. Now, a bootstrap analysis will be performed on the other two age groups' data to determine if the drug use ratio of 18-25 year olds could come from the distributions of the other age groups. 

Hypothesis: The ratio of use for 18-25 year olds in 2018 was significantly higher than the ratios of use for the other age groups. 

```python
## methods to be used to perform a bootstrap analysis 
def bootstrap(data, num): # bootstrapping method 
    samples = np.zeros(num) # empty array to be filled with sample counts 
    for ii in range(num): 
        # chooses sample from data with replacement
        sample = np.random.choice(data, size=len(data), replace=True) 
        count = np.sum(sample) # counts number of 1s in sample 
        samples[ii] = count # saves sample count to array 
    return samples # returns array of counts 

def confidenceIntervals(dataArray,pValue=0.05): # method to get exact confidence interval  
    sorted = np.sort(dataArray) # sort array 
    lowIndex = int((pValue/2)*len(sorted))-1  # get low end cut-off value
    highIndex = int((1-(pValue/2))*len(sorted))-1   # get high end cut-off value
    return sorted[lowIndex], sorted[highIndex] # returns exact cut-off values 

def tailsHisto(d1, binWidth, pValue, ageGroup): # function to graph bootstrapped data with confidence interval
    nB = len(d1)
    (n, bins, p) = plt.hist(d1, bins=np.arange(min(d1), max(d1) + binWidth, binWidth), edgecolor = 'black') # plot in histogram 
    sum = 0
    for i in range(0,len(p)): # changing color of out of confidence interval values 
        sum += n[i]
        if sum < nB*(pValue/2): 
            p[i].set(color = 'r')
            p[i].set(edgecolor='black')
        if sum > nB*(1-(pValue/2)): 
            p[i].set(color = 'r')
            p[i].set(edgecolor='black')
    plt.ylabel('Frequency')
    plt.xlabel('Sample Ratio of Users Per 1,530,000 People')
    plt.title(('Bootstrapped Distribution for Drug Use in 2018 of '+ ageGroup +' Year Olds'))
    plt.show()       
```

```python
numOfBootstrappedCounts = 1000 # number of samples to take 

props = bootstrap(virtualG1, numOfBootstrappedCounts) # get sample counts by bootstrapping 
tailsHisto(props, 13, .05, '12-17') # plot distribution with CI cutoff values 
low, high = confidenceIntervals(props) # get exact confidence interval values 
print('5% Confidence Interval Cutoff Values:', low, high)

props = bootstrap(virtualG3, numOfBootstrappedCounts) # get sample counts by bootstrapping 
tailsHisto(props, 14, .05, '26+') # plot distribution with CI cutoff values 
low, high = confidenceIntervals(props) # get exact confidence interval values 
print('5% Confidence Interval Cutoff Values:', low, high)
```

FIGURES 2-3: The above 2 plots are the sampling distributions created for two age groups by bootstrapping their drug use proportion data from 2018. The age group represented by each plot is specified in the plot title.
